```python
def reduzirPraUmDigito(numero):
    """Versão recursiva para redução a um dígito"""
    if numero < 10:
        return numero
    return reduzirPraUmDigito(sum(int(d) for d in str(numero)))
```

```python
def gerarPiramide(nomeEmString):
    """
    Processa a string e retorna o histórico completo de todas as listas.
    
    Args:
        string_numeros: String contendo dígitos
    
    Returns:
        Lista de listas com todo o histórico das atribuições
    """
    listaAtual = [int(c) for c in nomeEmString]
    piramide = [listaAtual.copy()]  # Armazena o primeiro valor
    
    for _ in range(len(nomeEmString)):
        if len(listaAtual) == 1:
            break
        
        # Criar nova lista
        listaAtual = [reduzirPraUmDigito(listaAtual[i] + listaAtual[i+1]) 
                       for i in range(len(listaAtual)-1)]
        
        # Armazenar cada nova atribuição
        piramide.append(listaAtual.copy())
    
    return piramide
```

```python
# Após executar o processo
nomeEmString = "79879797"
piramide2Print = gerarPiramide(nomeEmString)

for i, lista in enumerate(piramide2Print):
        print(f"  Sequência {i+1}: {lista}\n")

print(f"\n  Tamanho da pirâmide: {len(piramide2Print)}")
```

```python
MapaCabalistico = {
    1: ["a", "i", "q", "j", "y", ""],
    2: ["b", "k", "r", "´", "", ""],
    3: ["c", "g", "l", "s", "~", "á"],
    4: ["d", "m", "t", "", "", ""],
    5: ["e", "h", "n", "", "", ""],
    6: ["u", "v", "w", "x", "ç", ""],
    7: ["o", "z", "^", "", "", ""],
    8: ["f", "p", "", "", "", ""]
}
```

```python
print(MapaCabalistico)
```

```python
# Criar dicionário reverso
reverso = {caractere: key for key, valores in MapaCabalistico.items() 
           for caractere in valores if caractere}
```

```python
print(reverso)
```

```python
def converter_string_para_lista(texto):
    """Converte uma string em uma lista de inteiros conforme o MapaCabalistico
    Remove espaços em branco antes da conversão"""
    # Remove espaços em branco da string
    texto_sem_espacos = texto.replace(" ", "")
    return [reverso.get(caractere.lower(), 0) for caractere in texto_sem_espacos]
```

```python
resultado = converter_string_para_lista("Mariá Dinniz")
```
